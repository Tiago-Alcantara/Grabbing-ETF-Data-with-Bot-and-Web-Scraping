# Passo 2: Importar as bibliotecas

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

# Passo 3 : Entender como funciona as reqisições na internet

```python
drive = webdriver.Chrome(service= Service(ChromeDriverManager().install()))


url = "https://www.etf.com/etfanalytics/etf-finder"

drive.get(url)
```

# Passo 4: Conhecer e mapear o processo de coleta de dados no site do ETF.com

# Processos de dados

* Abrir o site 
* Mudar a visualização para 100
* Ler a tabela 
* Avançar todas as paginas
* Trocar para outra categoria
* Ler todas as tabelas de todas as categorias

# Passo 5.1 - Achar todos os elementos necessarios dentro do html do site - Pegando os numeros de paginas

```python
time.sleep(5)

botao_100= drive.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]
                                /div[2]/section[2]/div[1]/div/div[4]/button/label/span''')

drive.execute_script("arguments[0].click();",botao_100)
```

# Passo 5.2 - Achar todos os elementos necessarios dentro do html do site - Avançar todas as paginas

```python
numero_paginas = drive.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/label[2]''')

numero_paginas = numero_paginas.text.replace("of ","")

numero_paginas = int(numero_paginas)

```

# Passo 6.1 - ler a tabela de dados - Lendo a tabela de dados basicos

```python

lista_de_tabelas_por_pagina = []


for pagina in range (0,numero_paginas):

    tabela = drive.find_element("xpath",'''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')

    html_tabela = tabela.get_attribute("outerHTML")

    tabela_final = pd.read_html(html_tabela)[0]

    lista_de_tabelas_por_pagina.append(tabela_final)

    botao_avanca_pagina = drive.find_element("xpath",'''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')

    drive.execute_script("arguments[0].click();",botao_avanca_pagina)

base_dados_completa = pd.concat(lista_de_tabelas_por_pagina)

display(base_dados_completa)
```

# Passo 6.2 - ler a tabela de dados - Lendo a tabela de dados rentabilidade

```python

#primeira coisa mudar a aba 

botao_aba = drive.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/ul/li[2]/span''')

drive.execute_script("arguments[0].click();",botao_aba)

#voltando a primeira pagina

for pagina in range (0,numero_paginas):

    botao_voltar_pagina = drive.find_element("xpath",'''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[1]''')

    drive.execute_script("arguments[0].click();",botao_voltar_pagina)

lista_de_tabelas_por_pagina = []

for pagina in range (0,numero_paginas):

    tabela = drive.find_element("xpath",'''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')

    html_tabela = tabela.get_attribute("outerHTML")

    tabela_final = pd.read_html(html_tabela)[0]

    lista_de_tabelas_por_pagina.append(tabela_final)

    botao_avanca_pagina = drive.find_element("xpath",'''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')

    drive.execute_script("arguments[0].click();",botao_avanca_pagina)

base_dados_performace = pd.concat(lista_de_tabelas_por_pagina)
```

```python
drive.quit()
```

```python
base_dados_performace
#base_dados_completa
```

# Passo 7 - Contruir a tabela final

```python
base_dados_completa = base_dados_completa.set_index('Ticker')

base_dados_performace = base_dados_performace.set_index('Ticker')
base_dados_performace = base_dados_performace[['1 Year', '5 Years','10 Years']]

#juntar as tabelas
base_dados_final = base_dados_completa.join(base_dados_performace)
```
